```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('BankChurners.csv')
```

```python
print("--- Dataset Information ---")
print(df.info())
```

```python
print("\n--- First 5 Rows ---")
print(df.head())
```

```python
cols_to_drop = [col for col in df.columns if 'Naive_Bayes' in col]
df = df.drop(columns=cols_to_drop)
```

```python
print("\n--- Missing Values Count ---")
print(df.isnull().sum())
```

```python
print("\n--- Descriptive Statistics ---")
print(df.describe())
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Attrition_Flag', data=df, palette='viridis')
plt.title('Distribution of Customer Attrition (Existing vs Attrited)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Customer_Age'], kde=True, color='blue')
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Attrition_Flag', data=df)
plt.title('Gender vs Attrition Analysis')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Card_Category', y='Credit_Limit', data=df)
plt.title('Credit Limit Distribution by Card Category')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
numerical_df = df.select_dtypes(include=['number'])
sns.heatmap(numerical_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```
